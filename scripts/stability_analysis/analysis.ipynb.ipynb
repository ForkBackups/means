Function to compile data from various .pickle files in the specified tarfile

```python
def compile_data_from_tar_gz(tar_filename):
    import tarfile
    import cPickle as pickle
    from collections import OrderedDict

    df = []

    with tarfile.open(tar_filename) as tf:
        for entry in tf:
            file_ = tf.extractfile(entry)
            if file_ is None:
                continue
                
            data = pickle.load(file_)

            kwargs = data['kwargs']
            parameters = kwargs.pop('parameters')
            initial_conditions = kwargs.pop('initial_conditions')

            row = OrderedDict()
            for symbol, parameter in zip(MODEL.constants, parameters):
                row[str(symbol)] = parameter

            for symbol, initial_condition in zip(MODEL.species, initial_conditions):
                row[str(symbol)] = initial_condition
            exception = data['exception']
            row['exception'] = exception
            for key, value in kwargs.iteritems():
                row[key] = value

            df.append(row)

    return pd.DataFrame(df)
```

Compile the data into a dataframe called data

```python
%%cache compiled-data.pkl data
data = compile_data_from_tar_gz('data.tar.gz')
```

```python
def plot_scatter(frame, columns=None, alpha=0.05, density=False):
    """
    This is similar to pandas.tools.plotting.scatter_matrix, but changed to plot two variables at once
    """
    
    if columns is not None:
        frame = frame[columns + ['exception']]
    iterable_columns = frame.columns - ['exception']
    frame_no_exception = frame[frame.exception.isnull()]
    frame_exception = frame[~frame.exception.isnull()]
    
    n = len(iterable_columns)
    
    def boundaries(range_, range_padding=0.3):
        padding = (range_[0] - range_[1]) * range_padding / 2.
        return range_[0] - padding, range_[1] + padding
    
    for i, y_col in enumerate(iterable_columns):
        for j, x_col in enumerate(iterable_columns):
            index = i*n + j + 1
            ax = plt.subplot(n, n, index)
            xrange_ = frame[x_col].min(), frame[x_col].max()
            if i == j:
                
                if density:
                    from scipy.stats import gaussian_kde
                    y = frame_no_exception[x_col].values
                    gkde = gaussian_kde(y)
                    ind = np.linspace(xrange_[0], xrange_[1], 1000)
                    ax.plot(ind, gkde.evaluate(ind), color='g')
                    
                    y = frame_exception[x_col].values
                    gkde = gaussian_kde(y)
                    ind = np.linspace(xrange_[0], xrange_[1], 1000)
                    ax.plot(ind, gkde.evaluate(ind), color='r')
                else:
                    plotting_pd = pd.DataFrame({'exception' : frame_exception[x_col], 
                                                'no_exception': frame_no_exception[x_col]})
                    plotting_pd.apply(pd.value_counts).plot(kind='bar', stacked=True, ax=ax,
                                                            color=['r', 'b'], legend=False)
                    #frame_no_exception[x_col].hist(histtype="bar", color="g", range=range_, ax=ax, bins=20)
                    #frame_exception[x_col].hist(histtype="bar", color="r", range=range_, ax=ax, bins=20)
                
                   
            else:
                yrange_ = frame[y_col].min(), frame[y_col].max()
                #ax.scatter(frame_no_exception[x_col], frame_no_exception[y_col], alpha=alpha, color='g')
                #ax.scatter(frame_exception[x_col], frame_exception[y_col], alpha=alpha, color='r')
                
                #plt.hexbin(frame_exception[x_col], frame_exception[y_col])
                
                histogram, x, y = np.histogram2d(frame_no_exception[x_col].values, frame_no_exception[y_col].values, 
                                                 range=[xrange_, yrange_],
                                                 normed=False)
                
                
                sizes = []
                points_x = []
                points_y = []
                for x_coord, h in zip(x, histogram):
                    for y_coord, val in zip(y, h):
                        if val > 0:
                            sizes.append(val*10)
                            points_x.append(x_coord)
                            points_y.append(y_coord)
                            
                plt.scatter(points_x, points_y, s=sizes, alpha=0.3, color='b')
                
                histogram, x, y = np.histogram2d(frame_exception[x_col].values, frame_exception[y_col].values, 
                                                 range=[xrange_, yrange_],
                                                 normed=False)
                
                
                sizes = []
                points_x = []
                points_y = []
                for x_coord, h in zip(x, histogram):
                    for y_coord, val in zip(y, h):
                        if val > 0:
                            sizes.append(val*10)
                            points_x.append(x_coord)
                            points_y.append(y_coord)
                            
                plt.scatter(points_x, points_y, s=sizes, alpha=0.3, color='r')
    
                ax.set_xlim(*boundaries(xrange_))
                ax.set_ylim(*boundaries(yrange_))
                
            
            
            ax.set_ylabel(y_col, visible=True)
            ax.yaxis.set_visible(True)
            ax.yaxis.set_ticks_position('left')
            ax.yaxis.set_label_position('left')
            
            
            
            ax.set_xlabel(x_col, visible=True)
            ax.xaxis.set_visible(True)
            ax.xaxis.set_ticks_position('bottom')
            ax.xaxis.set_label_position('bottom')
            plt.setp(ax.get_xticklabels(), rotation=90)
    
            if j != 0 and not (j == 1 and i==0):
               ax.yaxis.set_visible(False)
            if i != n-1:
               ax.xaxis.set_visible(False)
                
    plt.gcf().subplots_adjust(wspace=0, hspace=0)
        
```
