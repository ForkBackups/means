Function to compile data from various .pickle files in the specified tarfile

```python
def compile_data_from_tar_gz(tar_filename):
    import tarfile
    import cPickle as pickle
    from collections import OrderedDict

    df = []

    with tarfile.open(tar_filename) as tf:
        for entry in tf:
            file_ = tf.extractfile(entry)
            if file_ is None:
                continue
                
            data = pickle.load(file_)

            kwargs = data['kwargs']
            parameters = kwargs.pop('parameters')
            initial_conditions = kwargs.pop('initial_conditions')

            row = OrderedDict()
            for symbol, parameter in zip(MODEL.constants, parameters):
                row[str(symbol)] = parameter

            for symbol, initial_condition in zip(MODEL.species, initial_conditions):
                row[str(symbol)] = initial_condition
            exception = data['exception']
            row['exception'] = exception
            for key, value in kwargs.iteritems():
                row[key] = value

            df.append(row)

    return pd.DataFrame(df)
```

Compile the data into a dataframe called data

```python
%%cache compiled-data.k5-k6-only.pkl data
data = compile_data_from_tar_gz('data-k5-k6-only.tar.gz')
```

```python
data_exceptions_as_status = data.copy()
data_exceptions_as_status.exception = ~data.exception.isnull()
```

```python
from ggplot import *
```

```python
def plot_scatter(data, x, y):
    data_exception = data[data.exception]
    data_no_exception = data[~data.exception]
    
    ax = plt.gca()
    ax.scatter(data_no_exception[x], data_no_exception[y], color='b', label='No Exception', alpha=0.4)
    ax.scatter(data_exception[x], data_exception[y], color='r', label='Exception', alpha=0.4)
    
    ax.set_xlabel(x)
    ax.set_ylabel(y)
    
       
    
```

```python
max_order = 2
plot_scatter(data_exceptions_as_status[data_exceptions_as_status.max_order==max_order], 'c_4', 'c_5')
plt.title('max_order = {0}'.format(max_order))
```
