```python
import means
import means.pipes
import means.examples
```

```python
class P53Model(means.Model):
    """
    A wrapper around means. Model that initialises it with P53 parameters and changes the __unicode__ function to
    print a shorter string
    """
    def __init__(self):
        from means.examples import MODEL_P53
        super(P53Model, self).__init__(MODEL_P53.species, MODEL_P53.parameters, MODEL_P53.propensities,
                                       MODEL_P53.stoichiometry_matrix)

    def __str__(self):
        # Override the str() methods so they do not print the whole blerch of things, but
        # only a nice and easily readable "p53"
        return 'p53'
```

```python
PARAMETERS_FROM_PAPER = [90.0, 0.002, 1.7, 1.1, 
                         0.9, # todo: change to 0.93, thanx 
                         0.96, 0.01]
INITIAL_CONDITIONS_FROM_PAPER = [70.0, 30.0, 60.0]
```

```python
ssa_trajectory_task = means.pipes.SSATrajectoryTask(model=P53Model(),
                                              parameters=PARAMETERS_FROM_PAPER,
                                              initial_conditions=INITIAL_CONDITIONS_FROM_PAPER,
                                              timepoints_arange=(0.0,40.0,0.1),
                                              n_simulations=5000)
```

```python
ssa_trajectories = ssa_trajectory_task.output().load()
```

```python
mea_task = means.pipes.MEATask(model=P53Model(),max_order=1,closure='scalar',multivariate=True)
```

```python
mea_problem = mea_task.output().load()
```

```python
inference_result = inference_mea.infer(return_intermediate_solutions=True)
```

```python
from matplotlib import pyplot as plt
from matplotlib.artist import setp
```

```python
def plot_trajectory(inference_result, parameter_x, parameter_y,legend=True, ax=None, start_end_only=False):
    if ax is None:
        ax = plt.gca()
        
    #parameter_number = len(inference_result.problem.parameters)
    index_x,index_y=None, None
    for i, parameter_name in enumerate(inference_result.problem.parameters):
        if parameter_name == parameter_x or str(parameter_name) == parameter_x:
            index_x = i
        if parameter_name == parameter_y or str(parameter_name) == parameter_y:
            index_y = i

    if index_x is None or index_y is None:
        raise KeyError('invalid parameter_x or parameter_y provided:' 
                       'x: {0!r}, y: {1!r}, available parameters: {2!r}'.format(parameter_x, parameter_y, 
                                                                                inference_result.problem.parameters))
    x, y = [], []
    for parameters, initial_conditions in inference_result.solutions:
        x.append(parameters[index_x])
        y.append(parameters[index_y])
    
    if not start_end_only:
        ax.plot(x,y)
    max_x = max(x)
    min_x = min(x)
    padding_x = (max_x - min_x) * 0.1 / 2.0
    
    max_y = max(y)
    min_y = min(y)
    padding_y = (max_y - min_y) * 0.1 / 2.0
    
    
    ax.set_xlim(min(x)-padding_x, max(x)+padding_x)
    ax.set_ylim(min(y)-padding_y, max(y)+padding_y)
    ax.plot(x[0], y[0], 'o', label='Start', color='b')
    ax.plot(x[-1], y[-1], 'x', label='End', color='r')
    
    ax.set_xlabel("${0}$".format(parameter_x), fontsize = 20)
    ax.set_ylabel("${0}$".format(parameter_y), fontsize = 20)
    setp(ax.get_xticklabels(),rotation=90)
    if legend:
        ax.legend()
    
    
    
```

```python
plot_trajectory(inference_result, 'c_1', 'c_3')
```

```python
fig= plt.figure(figsize=(20,20))
fig.subplots_adjust(wspace=0, hspace=0)
dimension = len(inference_result.problem.parameters)
for i,parameter_y in enumerate(inference_result.problem.parameters):
    for j, parameter_x in enumerate(inference_result.problem.parameters):
        
        ax = plt.subplot(dimension,dimension, i*dimension + j + 1)
        plot_trajectory(inference_result,parameter_x,parameter_y,legend=False, ax=ax, start_end_only=i==j)
        if j != 0:
            ax.yaxis.set_visible(False)

        if i != dimension - 1:
            ax.xaxis.set_visible(False)
        
        
```
