```python
import means
import means.pipes
import means.examples
```

```python
class P53Model(means.Model):
    """
    A wrapper around means. Model that initialises it with P53 parameters and changes the __unicode__ function to
    print a shorter string
    """
    def __init__(self):
        from means.examples import MODEL_P53
        super(P53Model, self).__init__(MODEL_P53.species, MODEL_P53.parameters, MODEL_P53.propensities,
                                       MODEL_P53.stoichiometry_matrix)

    def __str__(self):
        # Override the str() methods so they do not print the whole blerch of things, but
        # only a nice and easily readable "p53"
        return 'p53'
```

```python
PARAMETERS_FROM_PAPER = [90.0, 0.002, 1.7, 1.1, 
                         0.9, # todo: change to 0.93, thanx 
                         0.96, 0.01]
INITIAL_CONDITIONS_FROM_PAPER = [70.0, 30.0, 60.0]
```

```python
ssa_trajectory_task = means.pipes.SSATrajectoryTask(model=P53Model(),
                                              parameters=PARAMETERS_FROM_PAPER,
                                              initial_conditions=INITIAL_CONDITIONS_FROM_PAPER,
                                              timepoints_arange=(0.0,40.0,0.1),
                                              n_simulations=5000)
```

```python
ssa_trajectories = ssa_trajectory_task.output().load()
```

```python
mea_task = means.pipes.MEATask(model=P53Model(),max_order=1,closure='scalar',multivariate=True)
```

```python
mea_problem = mea_task.output().load()
```

```python
mea_problem
```

```python
inference_mea = means.Inference(mea_problem, 
                                PARAMETERS_FROM_PAPER, 
                                INITIAL_CONDITIONS_FROM_PAPER,
                                variable_parameters=mea_problem.parameters, #{'c_0': ()'c_2':(0,10),'c_4':(0,10)},
                                observed_trajectories=ssa_trajectories,
                                distance_function_type='sum_of_squares')
```

```python
inference_result = inference_mea.infer(return_intermediate_solutions=True, return_distance_landscape=True)
```
