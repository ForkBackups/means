```python
import pandas as pd
import numpy as np
```

```python
runtimes = pd.read_csv("runtimes.csv", index_col=0)
```

```python
def scatter_with_log_regression_line(df, x_col, y_col, log_scale=True, ylim=None, ax=None, *args, **kwargs):
    if ax is None:
        ax = plt.gca()
    
    if log_scale:
        ax.set_yscale('log')
    if ylim is not None:
        ax.set_ylim(*ylim)
        
    label = kwargs.pop('label', '')
    # Make lines the same color
    colour = kwargs.pop('color', 'r')
    alpha = kwargs.pop('alpha', 1)
    
    x = df[x_col]
    y = df[y_col]
    # Note no label parameter, we dont want these for scatter
    ax.scatter(x, y, color=colour, alpha=alpha, *args, **kwargs)
        
    # Fit a line trough log of the datapoints
    polynomial_fit = np.poly1d(np.polyfit(x, np.log(y), 1))
    regression_line = lambda v: np.exp(polynomial_fit(v))
    # Note no alpha parameter, we don't want these for lines
    ax.plot(x, regression_line(x), color=colour, label=label, *args, **kwargs)
    
```

```python
import itertools
def scatter_for_all_parameter_sets(df, log_scale=True, title=None, ylim=None, *args, **kwargs):
    colours = itertools.cycle(['b', 'r', 'g', 'y', 'k'])
    distinct_parameters = set(df.parameter_set.values)
    ax = plt.gca()
     
    
        
    for parameter in sorted(distinct_parameters):
        colour = colours.next()
        df_parameter = df[df.parameter_set == parameter]
        scatter_with_log_regression_line(df_parameter, 'max_order', 'runtime', 
                                         color=colour, label=parameter, ax=ax,
                                         log_scale=log_scale, ylim=ylim)
        
    
    plt.xlabel('max_order')
    plt.ylabel('runtime (seconds)')
    if title is not None:
        plt.title(title)
   
        
    plt.legend(loc='upper left', bbox_to_anchor=(0, 1))
```

```python
GROUP_COLUMNS = list(runtimes.columns - ['runtime', 'parameter_set', 'max_order', 'closure'])
# Make sure solver is the first column to group by
solver_index = GROUP_COLUMNS.index('solver')
GROUP_COLUMNS = ['solver'] + GROUP_COLUMNS[:solver_index] + GROUP_COLUMNS[solver_index+1:]
GROUP_COLUMNS
```

```python
max_min_diff = runtimes.runtime.max() - runtimes.runtime.min()
padding = max_min_diff * 0.1
ylim = (runtimes.runtime.min(), runtimes.runtime.max())
ylim = (ylim[0] -padding, ylim[1] + padding)
```

```python
def _groups(df):
    return runtimes.replace(np.nan, 'N/A').groupby(GROUP_COLUMNS)

def _str_key(key):
    named_key = zip(GROUP_COLUMNS, key)
    named_key = filter(lambda x: x[1] != 'N/A', named_key)
    
    
    title = '; '.join([ '{0}={1!r}'.format(key, value) for key, value in named_key])
    
    return title
```

```python
for key, group in runtimes.replace(np.nan, 'N/A').groupby(GROUP_COLUMNS):
    title = _str_key(key)
    closures = set(group.closure.values)
    plt.figure()
    
    for i, closure in enumerate(sorted(closures)):
        plt.subplot(1, len(closures), i+1)
        plt.title(closure)
        
        scatter_for_all_parameter_sets(group[group.closure==closure], ylim=ylim)
    
    plt.suptitle(title)
    if SVG_OUTPUT:
        plt.savefig(os.path.join(SVG_OUTPUT_DIR, '{0}.svg'.format('.'.join([str(x) for x in key if x != 'N/A']))))
        
    
        
```

```python
available_parameter_sets = set(runtimes.parameter_set.values)

colours = itertools.cycle(['b', 'g', 'r', 'c', 'm', 'y', 'k'])

colormap = plt.cm.spectral
colours = [colormap(i) for i in np.linspace(0, 0.9, len(_groups(runtimes)))]

for colour, (key, group) in zip(colours, _groups(runtimes)):
    
    title = _str_key(key)
    for i, parameter in enumerate(sorted(available_parameter_sets)):
        plt.subplot(1, len(available_parameter_sets), i+1)
        plt.title(parameter)
        
        sub_group = group[group.parameter_set == parameter]
        
        scatter_with_log_regression_line(sub_group, 'max_order', 'runtime', label=title, color=colour, alpha=0.1,
                                         ylim=ylim)
        plt.ylabel('Time (s), log scale')
        plt.xlabel('Max order')
    
    # Only one legend
    plt.legend(loc='lower left', bbox_to_anchor=(1, 0))
    plt.suptitle('Solver Runtimes')

if SVG_OUTPUT:
    plt.savefig(os.path.join(SVG_OUTPUT_DIR, 'runtimes-all.svg'), bbox_inches='tight')
    
```
