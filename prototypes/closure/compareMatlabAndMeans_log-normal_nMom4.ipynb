The huge difference is unexpected: let us try to compare equations in an exact way:

```python
for m,s in zip(problem_matlab.right_hand_side, problem_sympy.right_hand_side):
    print sympy_expressions_equal(m,s)
```

still the same! Let us try to simplify expr

```python
new_rhs = problem_sympy.right_hand_side.applyfunc(lambda x: sympy.simplify(sympy.expand(x)))
```

```python
# make a new problem with simplkified exprs:
problem_sympy_simple = ODEProblem("MEA", problem_sympy.ode_lhs_terms , new_rhs,  problem_sympy.constants)
```

```python
simulator_sympy_simple = means.simulation.Simulation(problem_sympy_simple)
trajectories_sympy_simple = simulator_sympy_simple.simulate_system(constant_values, initial_conditions, timepoints)
```

```python
plt.figure()
plt.plot(trajectories_matlab[0].values, trajectories_sympy_simple[0].values)
plt.title("Correlation of first speacies mean")
plt.show()

for tm,sm, sms in zip(trajectories_matlab, trajectories_sympy, trajectories_sympy_simple):
    plt.figure()
    tm.plot(label="matlab expr")
    sm.plot(label="sympy expr")
    sms.plot(label="sympy simple expr")
    plt.title("Moment compared: {0}".format(tm.description.symbol))
    plt.legend(loc='lower left', bbox_to_anchor = (1, 0))
    plt.show()
```
