```python
import means
from means.examples.sample_models import *
```

```python
import numpy as np
```

```python
import sympy as sp
```

```python
model = MODEL_P53
```

```python
n_moments = 3
problem_zero = means.approximation.MomentExpansionApproximation(model, n_moments,None).run()
```

```python
problem_gamma = means.approximation.MomentExpansionApproximation(model, n_moments, "gamma",1).run()
```

```python
for l,r in zip(problem_gamma.left_hand_side, problem_gamma.right_hand_side):
    print (l,[j for j in r.atoms() if isinstance(j, sp.Symbol)])
```

```python
# Values used in Ale et al.
constant_values = [90, 0.02, 1.7, 1.1, 0.93, 0.96, 0.01] 
initial_conditions = [70, 50, 60]
timepoints = np.arange(0, 20, 0.1)
```

```python
simulator_gamma = means.simulation.Simulation(problem_gamma)
__, trajectories_gamma = simulator_gamma.simulate_system(constant_values, initial_conditions, timepoints)
```

```python
simulator_zero = means.simulation.Simulation(problem_zero)
__, trajectories_zero = simulator_zero.simulate_system(constant_values, initial_conditions, timepoints)

```

```python
~problem_gamma.right_hand_side - problem_zero.right_hand_side
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure()

plt.plot(trajectories_zero[5].timepoints,trajectories_zero[5].values)
plt.plot(trajectories_gamma[5].timepoints,trajectories_gamma[5].values)
for i in range(len(trajectories_zero)):
    plt.figure()

    dist = sum((trajectories_zero[i].values -  trajectories_gamma[i].values))
    print  "Distance = {0}".format(dist)
    
    
    plt.plot(trajectories_zero[i].values,  trajectories_gamma[i].values)
    

```
