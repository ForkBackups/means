```python
import numpy as np
import means
import pylab as pl
from means.examples.sample_models import *
```

```python
birth_death_model = means.Model(parameters=['b', 'd'],
                   species=['Cell'],
                   propensities=['Cell * b',
                                 'd * Cell * (Cell-1)/2'],
                   stoichiometry_matrix=[[1, -2]])
                                         
```

```python
problem = means.mea_approximation(birth_death_model,1)
simulation = means.Simulation(problem)
timerange = np.arange(0,50,0.1)
bd_trajs = simulation.simulate_system([.2,1e-3],[5], timerange)
bd_trajs

fig = pl.figure()
bd_trajs.plot()
fig.savefig("/tmp/cell_growth_determ.png",dpi=500)
```

```python
problem = means.mea_approximation(MODEL_LOTKA_VOLTERRA,1)
simulation = means.Simulation(problem)
timerange = np.arange(0,40,0.1)
params = [0.4,0.01,0.5]
t0 = [50,50]
lv_trajs = simulation.simulate_system(params,t0, timerange)
lv_trajs
fig = pl.figure()
lv_trajs.plot()
fig.savefig("/tmp/lv_determ.png",dpi=500)
```

```python
problem = means.StochasticProblem(MODEL_LOTKA_VOLTERRA)
simulation = means.SSASimulation(problem, 1,4)
lv_trajs_ssa = simulation.simulate_system(params,t0,timerange)
lv_trajs_ssa

fig = pl.figure()
lv_trajs_ssa.plot()
fig.savefig("/tmp/lv_one_SSA.png",dpi=500)
```

```python
problem = means.StochasticProblem(MODEL_LOTKA_VOLTERRA)
simulation = means.SSASimulation(problem, 30,1)
lv_trajs_ssa = simulation.simulate_system(params,t0,timerange,0)
```
