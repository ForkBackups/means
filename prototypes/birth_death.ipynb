```python
import numpy as np
import means
import pylab as pl
from means.examples.sample_models import *
```

```python
birth_death_model = means.Model(parameters=['b', 'd'],
                   species=['Cell'],
                   propensities=['Cell * b',
                                 'd * Cell * (Cell-1)/2'],
                   stoichiometry_matrix=[[1, -2]])
                                         
```

```python
problem = means.mea_approximation(birth_death_model,1)
simulation = means.Simulation(problem)
timerange = np.arange(0,50,0.1)
bd_trajs = simulation.simulate_system([.2,1e-3],[5], timerange)
bd_trajs

fig = pl.figure()
bd_trajs.plot()
fig.savefig("/tmp/cell_growth_determ.png",dpi=500)
```

```python
problem = means.mea_approximation(MODEL_LOTKA_VOLTERRA,1)
simulation = means.Simulation(problem)
timerange = np.arange(0,40,0.1)
params = [0.4,0.01,0.5]
t0 = [50,50]
fig = pl.figure()
lv_trajs = simulation.simulate_system(params,t0, timerange)
lv_trajs.plot(legend=False)
pl.title("")
pl.ylabel("# of individuals")
fig.savefig("/tmp/lv_determ.svg", tight_layout=True)
fig.savefig("/tmp/lv_determ.pdf", tight_layout=True)
```

```python
problem = means.StochasticProblem(MODEL_LOTKA_VOLTERRA)
simulation = means.SSASimulation(problem, 1,4)
lv_trajs_ssa = simulation.simulate_system(params,t0,timerange)
lv_trajs_ssa

fig = pl.figure()
lv_trajs_ssa.plot(legend=False)
pl.title("")
pl.ylabel("# of individuals")
fig.savefig("/tmp/lv_one_SSA.png",dpi=500, tight_layout=True)
```

```python
problem = means.StochasticProblem(MODEL_LOTKA_VOLTERRA)
simulation = means.SSASimulation(problem, 100,1)
lv_trajs_ssa = simulation.simulate_system(params,t0,timerange,0)
```

```python
timepoint = 15
fig = pl.figure(figsize=(15, 5))
ax = pl.subplot(1, 2, 1)
max_ = 0
for t in lv_trajs_ssa[0:10]:
    print t[0].plot()
    max_ = max(max_, max(t[0].values))

pl.vlines(timepoint, 0, max_, color='k', linewidth=3, linestyles='dashed')
pl.xlabel('time')
plt.ylabel('# of individuals')
pl.subplot(1, 2, 2, sharey=ax)
h = [t[0].values[timepoint*10] for t in lv_trajs_ssa]
x, bins = np.histogram(h,8)
n, bins, patches = pl.hist(h, bins=20, normed=1, histtype='stepfilled', orientation='horizontal')
pl.hlines(np.mean(h), 0, 0.032, color='k', linestyles='dashed', linewidth=3)
pl.xlabel('Ratio')

fig.savefig("/tmp/lv_many_SSA.png",dpi=500)
```
