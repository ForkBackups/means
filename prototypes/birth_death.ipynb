```python
import numpy as np
import means
import pylab as pl
from means.examples.sample_models import *
```

```python
birth_death_model = means.Model(parameters=['b', 'd'],
                   species=['Cell'],
                   propensities=['Cell * b',
                                 'd * Cell * (Cell-1)/2'],
                   stoichiometry_matrix=[[1, -2]])
                                         
```

```python
problem = means.mea_approximation(birth_death_model,1)
simulation = means.Simulation(problem)
timerange = np.arange(0,50,0.1)
bd_trajs = simulation.simulate_system([.2,1e-3],[5], timerange)
bd_trajs
```

```python
problem = means.mea_approximation(MODEL_LOTKA_VOLTERRA,1)
simulation = means.Simulation(problem)
timerange = np.arange(0,35,0.1)
lv_trajs = simulation.simulate_system([0.25,0.1,0.8],[20,20], timerange)
lv_trajs
```

```python
problem = means.StochasticProblem(MODEL_LOTKA_VOLTERRA)
simulation = means.SSASimulation(problem, 1,2)
lv_trajs_ssa = simulation.simulate_system([0.25,0.1,0.8],[20,20],np.arange(0,10,0.1))
lv_trajs_ssa
```

```python
simulation = means.SSASimulation(problem, 1,1)
lv_trajs_ssa = simulation.simulate_system([0.25,0.1,0.8],[20,20],np.arange(0,10,0.1))
lv_trajs_ssa
```

```python
problem = means.StochasticProblem(MODEL_LOTKA_VOLTERRA)
simulation = means.SSASimulation(problem, 1000)
lv_trajs_ssa = simulation.simulate_system([0.25,0.1,0.8],[20,20], timerange)
lv_trajs_ssa
```
