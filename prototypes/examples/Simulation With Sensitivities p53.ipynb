```python
import means
import means.examples
import numpy as np
import matplotlib.pyplot as plt
import sympy
from itertools import cycle
```

```python
model = means.examples.sample_models.MODEL_P53
model
```

```python
parameters = [90, 0.002, 1.7, 1.1, 0.93, 0.96, 0.01]
initial_conditions = [70, 30, 60]
timepoints = np.arange(0, 100, 0.01)
```

```python
MOMENT_ORDER = 3
```

```python
problem = means.approximation.MomentExpansionApproximation(model, MOMENT_ORDER).run()
```

```python
simulation = means.simulation.SimulationWithSensitivities(problem)
```

```python
trajectories = simulation.simulate_system(parameters, initial_conditions, timepoints)
```

```python
for trajectory in trajectories:
    if trajectory.description.order != 1:
        continue
    COLORS = cycle(['b', 'g', 'r', 'c', 'm', 'y', 'k'])    
    plt.figure()
    trajectory.plot()
    for constant, parameter in zip(model.constants, parameters):
        r_delta = 0.1
        trajectory.plot_perturbations(constant, delta=r_delta*parameter, color=COLORS.next(),
                                      label='${0} = (1 \pm {1}){0}$'.format(constant, r_delta))
    plt.title('${0}$'.format(trajectory.description.symbol))
    plt.legend(loc='lower left', bbox_to_anchor = (1, 0))       
```

```python
plt.close()
```
