```python
# Values used in Ale et al.
constant_values = [90, 0.002, 1.7, 1.1, 0.93, 0.96, 0.01] 
initial_conditions = [70, 30, 60]
```

```python
timepoints = np.arange(0, 50)
```

```python
problems = []
for n_moments in range(1, 4):
    problem = means.approximation.MomentExpansionApproximation(model, n_moments).run()
    problems.append(problem)
```

```python
for problem in problems:
    print problem.right_hand_side
    print '----'
```

```python
results = []
for problem in problems[2:]:
    simulator = means.simulation.Simulation(problem)
    __, trajectories = simulator.simulate_system(constant_values, initial_conditions, timepoints)
    print trajectories[0]
    results.append(trajectories)
```
