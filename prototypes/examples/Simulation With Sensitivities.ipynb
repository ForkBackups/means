```python
import means
import means.examples
import numpy as np
import matplotlib.pyplot as plt
import sympy
from itertools import cycle
```

```python
model = means.model.Model(['k_1', 'k_2'], ['x_1', 'x_2'], ['k_1*x_1*(x_1-1)', 'k_2*x_2'], [[-2, 2], [1, -1]])
```

```python
parameters = [1.66e-3, 0.2]
initial_conditions = [301, 0]
timepoints = np.arange(0, 20, 0.01)
```

```python
MOMENT_ORDER = 3
```

```python
problem = means.mea_approximation(model, MOMENT_ORDER)
```

```python
problem
```

```python
simulation = means.simulation.SimulationWithSensitivities(problem)
```

```python
trajectories = simulation.simulate_system(parameters, initial_conditions, timepoints)
```

```python
COLORS = cycle(['b', 'g', 'r', 'c', 'm', 'y', 'k'])
for trajectory in trajectories:
    if trajectory.description.order != 1:
        continue
        
    plt.subplot(3,1,1)
    trajectory.plot()
    trajectory.plot_perturbations(sympy.Symbol('k_1'), delta=0.001, color=COLORS.next())
    plt.title('Means')
    plt.legend(loc='lower left', bbox_to_anchor = (1, 0))
    
    for i, sensitivity_trajectory in enumerate(trajectory.sensitivity_data):
        plt.subplot(3,1,2+i)
        plt.title('Sensitivities {0}'.format(sensitivity_trajectory.description.parameter))
        sensitivity_trajectory.plot()
        plt.legend(loc='lower left', bbox_to_anchor = (1, 0))
    
        
```

```python
plt.close()
```
