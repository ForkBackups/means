```python
model = means.examples.MODEL_P53
```

```python
MAX_ORDERS = [2, 3, 4, 5, 6]
```

```python
%%cache problems.pickle problems
problems = {}
for order in MAX_ORDERS:
    print 'Computing problem for {0!r}'.format(order)
    problems[order] = means.approximation.MomentExpansionApproximation(model, max_order=order).run()
```

```python
PARAMETERS_FROM_PAPER = [90, 0.002, 1.7, 1.1, 0.93, 0.96, 0.01]
INITIAL_CONDITIONS_FROM_PAPER = [70, 30, 60]
```

```python
timepoints = np.arange(0, 40, 0.1)
```

```python
def plot_only_first_order_trajectories(trajectories, *args, **kwargs):
    for trajectory in trajectories:
        if trajectory.description.order != 1:
            continue
        trajectory.plot(*args, **kwargs)
```

```python
%%cache trajectories.pickle computed_trajectories
computed_trajectories = {}
for order in MAX_ORDERS:
    s = means.simulation.Simulation(problems[order])
    computed_trajectories[order] = s.simulate_system(PARAMETERS_FROM_PAPER, INITIAL_CONDITIONS_FROM_PAPER, timepoints)
```

```python
for order, trajectories in computed_trajectories.iteritems():
    plt.figure()
    plot_only_first_order_trajectories(trajectories)
    plt.title(order)
```

```python
OSCILLATING_TRAJECTORIES = filter(lambda x: x.description.order == 1, computed_trajectories[6])
```

```python
ALL_PARAMETERS = model.constants + model.species
```

```python
def infer_with_lower_order(lower_order, observed_trajectories, 
                           variable_parameters=ALL_PARAMETERS, intermediate_solutions=False):
    inference = means.inference.inference.Inference(problems[lower_order], 
                                                    PARAMETERS_FROM_PAPER, 
                                                    INITIAL_CONDITIONS_FROM_PAPER,
                                                    variable_parameters=variable_parameters,
                                                    observed_trajectories=observed_trajectories,
                                                    method='sum_of_squares'
                                                    )
    result = inference.infer(return_intermediate_solutions=intermediate_solutions)
    return result
```

```python
result = infer_with_lower_order(2, OSCILLATING_TRAJECTORIES, intermediate_solutions=True)
result.plot(filter_plots_function=lambda x: x.description.order == 1)
result
```

```python
PARAMETERS_RANGE = [(70, 110), 
                    (0.001, 0.003),
                    (1.2, 2.2),
                    (0.6, 1.6),
                    (0.8, 2),
                    (0.8, 2),
                    (0.005, 0.015)]
INITIAL_CONDITIONS_RANGE = [(60, 80), (20, 40), (50, 70)]
```

```python
import multiprocessing
NUMBER_OF_PROCESSES = multiprocessing.cpu_count()
```

```python
def infer_with_lower_order_with_restarts(lower_order, observed_trajectories, number_of_restarts):
   
    number_of_processes = NUMBER_OF_PROCESSES
    inference = means.inference.inference.InferenceWithRestarts(problems[lower_order], number_of_restarts,
                                                                PARAMETERS_RANGE, 
                                                                INITIAL_CONDITIONS_RANGE,
                                                                variable_parameters=ALL_PARAMETERS,
                                                                observed_trajectories=observed_trajectories,
                                                                method='sum_of_squares'
                                                                )
    result = inference.infer(number_of_processes=number_of_processes)
    return result
```

```python
result = infer_with_lower_order_with_restarts(2, OSCILLATING_TRAJECTORIES, NUMBER_OF_PROCESSES*2)
result
```

```python
def generate_trajectories(parameters, initial_conditions):
    
    import means.simulation.ssa
    ssa_simulation = means.simulation.ssa.SSASimulator(model)
    
    # Yep this is infinite, be smart on how you use this function
    while True:
        trajectories = ssa_simulation.simulate_system(parameters, initial_conditions, timepoints[-1])
        trajectories = [t.resample(timepoints) for t in trajectories]
        yield trajectories
        
    
```

```python
N_TRAJECTORIES = 5000
```

```python
import itertools
```

```python
paper_trajectory_generator = generate_trajectories(PARAMETERS_FROM_PAPER, INITIAL_CONDITIONS_FROM_PAPER)
def one_paper_trajectory(x):
    return paper_trajectory_generator.next()

lower_moments_best_trajectory_generator = generate_trajectories(result.best.optimal_parameters, 
                                                                result.best.optimal_initial_conditions[:3])
def one_lower_moments_trajectory(x):
    return lower_moments_best_trajectory_generator.next()
```

```python
%%time
%%cache ssa-trajectories.pickle ssa_original_trajectories ssa_lower_moments_trajectories
pool = multiprocessing.Pool(NUMBER_OF_PROCESSES)
ssa_original_trajectories = pool.map(one_paper_trajectory, range(N_TRAJECTORIES))
ssa_lower_moments_trajectories = pool.map(one_lower_moments_trajectory, range(N_TRAJECTORIES))
```

```python
def mean_trajectories(trajectories_list):
    sum_trajectories = {}
    for trajectories in trajectories_list:
        for trajectory in trajectories:
            try:
                sum_trajectories[trajectory.description] += trajectory
            except KeyError:
                sum_trajectories[trajectory.description] = trajectory
        
    mean_trajectories = []
    for trajectory in sum_trajectories.values():
        trajectory = means.simulation.Trajectory(trajectory.timepoints, 
                                                 trajectory.values / float(N_TRAJECTORIES),
                                                 trajectory.description)
    
        mean_trajectories.append(trajectory)    
    
    return sorted(mean_trajectories, key=lambda x: str(x.description))
            
ssa_original_mean = mean_trajectories(ssa_original_trajectories)
ssa_lower_moments_mean = mean_trajectories(ssa_lower_moments_trajectories)
```

```python
for original_trajectory, lower_moment_trajectory in zip(ssa_original_mean, ssa_lower_moments_mean):
    plt.figure()
    
    assert(original_trajectory.description == lower_moment_trajectory.description)
    plt.subplot(1, 2, 1)
    plt.suptitle(original_trajectory.description)
    original_trajectory.plot(label="Original")
    lower_moment_trajectory.plot(label="Lower_moment")
    plt.legend()
    plt.subplot(1,2,2)
    diff_values = np.cumsum(np.square(original_trajectory.values - lower_moment_trajectory.values))
    plt.plot(original_trajectory.timepoints, diff_values, label='Cumulative sq. difference')
    plt.legend()
```
