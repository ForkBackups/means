```python
from ggplot import *
import pandas as pd
```

```python
import means
import means.examples
```

```python
from means.examples.sample_models import MODEL_P53
```

```python
model = MODEL_P53
```

```python
problem = means.approximation.MomentExpansionApproximation(model, 3).run()
```

```python
simulator = means.simulation.SimulationWithSensitivities(problem)
```

```python
parameters = [90, 0.002, 1.7, 1.1, 0.93, 0.96, 0.01]
initial_conditions = [70,30,60]
timepoints = np.arange(0, 40, 0.1)
```

```python
trajectories = simulator.simulate_system(parameters, initial_conditions, timepoints)
```

```python
trajectory_series = pd.Series(trajectories[0].values, index=trajectories[0].timepoints, name=trajectories[0].description)
```

```python
trajectory_series
```

```python
trajectory_series.name
```

```python
trajectories_df = pd.DataFrame({trajectory.description: pd.Series(trajectory.values, index=trajectory.timepoints) for trajectory in trajectories})
```

```python
melted_df = pd.melt(trajectories_df.reset_index(), id_vars=['index'])
melted_df
```

```python
ggplot(melted_df, aes(x='index', y='value')) + facet_wrap("variable") + geom_line()
```

```python
ggplot(melted_df[melted_df.variable.map(lambda x: x.order) == 1], aes(x='index', y='value', colour='variable')) + geom_line()
```
