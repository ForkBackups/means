```python
model = Model.from_file('dimerisation.model')
```

```python
model
```

```python
approximation = mea_approximation(model, 2)
approximation
```

```python
import means.io
data = means.io.from_file('dimerisation.data')
data = [data[0]] # Take only the mean
```

```python
inference = InferenceWithRestarts(approximation,
                                  number_of_samples=20,
                                  starting_parameter_ranges=[(0, 0.001), (0, 0.5), (260, 330)],
                                  starting_conditions_ranges=[(290, 320)],
                                  variable_parameters = {'c_0': (0, 0.001),
                                                         'c_1': (0, 0.5),
                                                         'c_2': (260, 330),
                                                         'y_0': (290, 320)},
                                  observed_trajectories=data,
                                  distance_function_type='normal',
                                  )
```

```python
result = inference.infer(number_of_processes=8)
```
